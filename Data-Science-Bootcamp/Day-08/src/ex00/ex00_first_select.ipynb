## Imports

```python
import pandas as pd
import sqlite3
```

## 1. Connection to the database. Schema of the pageviews table

```python
conn = sqlite3.connect('../data/checking-logs.sqlite')

schema = pd.io.sql.read_sql('PRAGMA table_info(pageviews);', conn)
print(schema)
```

## 2. Preview first 10 rows of the pageviews table

```python
first_rows = pd.io.sql.read_sql('SELECT * FROM pageviews LIMIT 10;', conn)
print(first_rows)
```

## 3. Filter user pageviews

```python
query = '''
    SELECT uid, datetime FROM pageviews
    WHERE uid LIKE 'user_%'
    ORDER BY uid ASC
'''

pageviews = pd.io.sql.read_sql(
    query,
    conn,
    index_col='datetime',
    parse_dates=['datetime']
)

conn.close()
```
